Generate 1000 random points.


Build a Linear Regression model from Scratch.


Use Gradient Descent of find best fit line (Slope, intercept) with leat squares criterion.


(with and without using sklearn)

```python
import pandas as pd
import numpy as np 
import random
import matplotlib.pyplot as plt
```

```python
np.random.seed(45)
x=np.random.randn(1000)
g=0.5
y=(g*np.random.randn(1000))+x
```

```python
print(x,y)
```

```python
def gd(LR,iterations,x,y):
    m=4
    c=0
    n = x.shape[0]
    for i in range(iterations): 
        Y = m*x + c 
        D_m = (-2/n) * np.sum(x * (y - Y)) 
        D_c = (-2/n) * np.sum(y - Y) 
        m = m - (LR * D_m  )
        c = c - (LR * D_c) 
    return m,c
```

```python
m,c=gd(0.2,1000,x,y)
Y_pred=m*x + c
plt.scatter(x,y,s=6,c='red')
plt.plot(x,Y_pred)
```

```python
y=y.reshape(-1,1)
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
X=reg.predict(y)
```

```python
plt.scatter(x,y,s=5,c='red')
plt.plot(y,X)
```

```python
print(" Using gradient Descend \n slope : ",m,"\t intercept : ",c)
print("\n Using Sci-Kit learn \n slope : ",reg.coef_[0],"\t intercept : ",reg.intercept_)
```
